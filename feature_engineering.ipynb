# missing values

```python
#categorical features-> 'O' dtype
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data=pd.read_csv('train.csv')
```

```python
#missing values in categorical features
missing_values_cat=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes=='O']
for feature in missing_values_cat:
    print('{}: {}% missing values'.format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
#replacing missing value with a new label
def replace_missing(data,missing_values):
    data2=data.copy()
    data2[missing_values]=data[missing_values].fillna('Missing')
    return data2
data=replace_missing(data,missing_values)
data[missing_values].isnull().sum()
```

```python
data.head()
```

```python
#missing values in numerical data
missing_values_num=[feature for feature in data.columns if data[feature].isnull().sum()>1 and data[feature].dtypes!='O']
for feature in missing_values_num:
    print('{} : {}% missing values'.format(feature,np.round(data[feature].isnull().mean(),4)))
```

```python
#replacing the numerical missing values with its median
#when there are lot of outliers we have to go with median or mode
for feature in missing_values_num:
    median_value=data[feature].median()
    
    #create new feature to capture nan values
    data[feature+'nan']=np.where(data[feature].isnull(),1,0)
    data[feature].fillna(median_value,inplace=True)
data[missing_values_num].isnull().sum()
```

```python
data.head(30)
```

```python
#temporal variable 
for feature in ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']:
    data[feature]=data['YrSold']-data[feature]
```

```python
data[['YearBuilt', 'YearRemodAdd', 'GarageYrBlt']].head()
```

```python
num_features=['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']
for feature in num_features:
    data[feature]=np.log(data[feature])
```

```python
data.head()
```

# handling rare categorical feature
we remove the categorical feature that are present in less than 1% of observations


```python
categorical_feature=[feature for feature in data.columns if data[feature].dtypes=='O']
categorical_feature
```

```python
#considering whole dataset find perc of each category
for feature in categorical_feature:
    temp=data.groupby(feature)['SalePrice'].count()/len(data)
    temp_df=temp[temp>0.01].index
    data[feature]=np.where(data[feature].isin(temp_df),data[feature],'Rare_var')
```

```python
for feature in categorical_feature:
    labels_ordered=data.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    data[feature]=data[feature].map(labels_ordered)
```

```python
data.head(10)
```

# Feature scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
feat_scal=[feature for feature in data.columns if feature not in ['Id','SalePrice']]
scaler=MinMaxScaler()
scaler.fit(data[feat_scal])
```

```python
# transform the train and test set, and add on the Id and SalePrice variables
data = pd.concat([data[['Id', 'SalePrice']].reset_index(drop=True),
                    pd.DataFrame(scaler.transform(data[feat_scal]), columns=feat_scal)],
                    axis=1)
```

```python
data.to_csv('train_final.csv',index=False)
```
